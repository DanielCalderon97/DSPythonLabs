##### 1) ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
#Usamos np.zeros para crear un array de 1000 elementos los cuales por defecto inician en False ya que en numpy False es equivalente a 0 y True a 1
array_booleano = np.zeros(1000, dtype=bool) # especificamos que el tipo de dato es booleano

#Devolvemos el total en bytes que ocupa el array en memoria
tamaño_bytes = array_booleano.nbytes #usamos .nbytes para obtener el tamaño en bytes del array

print(f"El tamaño en bytes del array es: {tamaño_bytes} bytes")
```

##### 3. Dado el array x Seleccione con una instrucción el subarray de elementos 0 3 7 4. Después, seleccione el subarray de elementos 2 5 || 8 5

```python
# Primero creamos el array brindado
array = np.array([
    [2, 4, 5, 6],
    [0, 3, 7, 4],
    [8, 8, 5, 2],
    [1, 5, 6, 1]
]) # Usamos separadores de [] y coma para separar los elementos de cada fila y de cada columna

#Validamos que se haya creado de manera correcta
print("El array original es:")
print(array)

#Seleccionamos los elementos de la segunda fila
segunda_fila = array[1,:] # Usamos : para seleccionar todos los elementos de la fila 1 (Ya que en Python los índices inician en 0)
# Se coloca ,: dentro del [] para indicar que se seleccionan todos los elementos de la fila 1
print("La segunda fila es:")
print(segunda_fila)

#Seleccionamos los elementos [2 5] y [8, 5]
elementos = array[[0,2], 0::2] # Se selecciona la fila 0 y la fila 2 con [0,2] y se selecciona desde la columna 0 hasta la última con  un paso de 2 usando el :: (la primera y la tercera en este caso)
# Se colocan los índices de las filas y columnas dentro de [] y separados por coma
print("Los elementos [2 5] y [8, 5] son:")
print(elementos)
```

##### 4. Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
#Primero creamos los arrays a y b
a = np.array([1,4,2,7])
b = np.array([1,3,2,9])

#Calculamos la diferencia entre los arrays
diferencia = a - b

#Calculamos el promedio de la diferencia dentro del array
promedio = np.mean(diferencia)

print(f"El promedio de la diferencia entre los arrays es: {promedio}")
```

##### 5. Tengo valores de cordenadas (x, y) en las columnas del array 1.33 || 4.5, 30.0 || 10.7, 70.2 || 0.5. Agregue a este array las coordenadas (37.1, -3.6). Muestre en pantalla las dimensiones del nuevo array.

```python
# Primero creamos el array brindado original
coordenadas = np.array([
    [1.33, 4.5],
    [30.0, 10.7],
    [70.2, 0.5]
])

# Ahora integramos las nuevas coordenadas al array original
nuevas_coordenadas = np.array([[37.1, -3.6]])
coordenadas = np.append(coordenadas, nuevas_coordenadas, axis=0) # Usamos np.append para agregar las nuevas coordenadas al array original, especificamos el eje en el que se agregarán las nuevas coordenadas con axis=0

print("El array con las nuevas coordenadas es:")
print(coordenadas)
```

##### 1) ¿Cuántos Municipios tienen más de 100,000 habitantes?

#### 2) Realice una gráfica de barras sobre la población de cada ciudad, ordenela de menor a mayor y responda:
 - a. ¿Cuál es la segunda ciudad más poblada?
 - b. ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

```python
# Primero ordenamos el dataset por población de manera descendente
dataset_ordenado = dataset.sort_values(by='Poblacion', ascending=False) # Usamos .sort_values para ordenar el dataset por la columna 'Poblacion' de manera descendente, especificamos que sea de manera descendente con ascending=False

# Creamos una grafica de barras con los 10 municipios con mayor población
plt.bar(dataset_ordenado['Municipio'][:10], dataset_ordenado['Poblacion'][:10]) # Usamos plt.bar para crear una gráfica de barras, especificamos que se grafiquen los primeros 10 municipios con mayor población
plt.xlabel('Municipio') # Usamos plt.xlabel para poner la etiqueta del eje x
plt.ylabel('Población') # Usamos plt.ylabel para poner la etiqueta del eje y
plt.xticks(rotation=90) # Usamos plt.xticks para rotar las etiquetas del eje x 90 grados para poder leerlas
plt.show() # Usamos plt.show para mostrar

# Averiguamos cual es la segunda ciudad más poblada
segunda_ciudad = dataset_ordenado.iloc[1]['Municipio'] # Usamos .iloc para seleccionar la segunda fila de la columna 'Municipio'
print(f"a. La segunda ciudad más poblada es: {segunda_ciudad}")

# Averiguamos la posición que ocupa Granada en el ranking de población
posicion_granada = dataset_ordenado[dataset_ordenado['Municipio'] == 'Granada'].index[0] # Usamos .index para obtener el índice de la fila que contiene 'Granada'
print(f"b. Granada ocupa la posición número: {posicion_granada}")
```

#### 3. ¿Cuántos municipios de La Rioja tienen más de 20,000 habitantes?.

```python
# Primero filtramos a La Rioja dentro de provincia
municipios_rioja = dataset[dataset['Provincia'] == 'La Rioja']
municiios_grandes_rioja = municipios_rioja[municipios_rioja['Poblacion'] > 20000] # Filtramos los municipios con población mayor a 20000 en 'La Rioja'
cuenta_municipios_grandes_rioja = len(municiios_grandes_rioja) # Contamos cuántos municipios hay con población mayor a 20000 en 'La Rioja'
print(f"El número de municipios con población mayor a 20000 en 'La Rioja' es: {cuenta_municipios_grandes_rioja}")
```

#### 4) ¿Cuál es el municipio situado más al Norte? (Usar el valor de la coordenada "Y" que representa la latitud en grados). Proporcione también la provincia a la que pertenece y su población.

```python
# Usamos .loc para seleccionar la fila que contiene la latitud más alta y usamos .idxmax para obtener el índice de la fila que contiene la latitud más alta
municipio_norte = dataset.loc[dataset['Y'].idxmax()] # Despues de obtener el índice de la fila que contiene la latitud más alta, seleccionamos la fila con .loc
print(f"El municipio más al norte es: {municipio_norte['Municipio']}") # Usamos el índice 'Municipio' para obtener el nombre del municipio más al norte
print(f"El municipio más al norte es: {municipio_norte['Poblacion']}") # Usamos el índice 'Poblacion' para obtener la población del municipio más al norte
print(f"El municipio más al norte es: {municipio_norte['Provincia']}")  # Usamos el índice 'Provincia' para obtener la provincia del municipio más al norte
```

##### 5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
# Filtramos los municipios de Granada a traves de la columna Provincia
municipios_granada = dataset[dataset['Provincia'] == 'Granada']
media_poblacion = municipios_granada['Poblacion'].mean() # Calculamos la media de la población de los municipios con mean()
mediana_poblacion = municipios_granada['Poblacion'].median() # Usamos median() para calcular la mediana de la población
desviacion_poblacion = municipios_granada['Poblacion'].std() # usamos std() para calcular la desviación estándar de la población
max_poblacion = municipios_granada['Poblacion'].max() # Calculamos la población máxima de los municipios de Granada
min_poblacion = municipios_granada['Poblacion'].min() # Calculamos la población mínima de los municipios de Granada

print(f"La media de la población de los municipios de Granada es: {media_poblacion}")
print(f"La mediana de la población de los municipios de Granada es: {mediana_poblacion}")
print(f"La desviación estándar de la población de los municipios de Granada es: {desviacion_poblacion}")
print(f"La población máxima de los municipios de Granada es: {max_poblacion}")
print(f"La población mínima de los municipios de Granada es: {min_poblacion}")
```

#### 6. Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
#Usamos for para recorrer cada provincia en el dataset
for provincia in dataset['Provincia'].unique(): # Usamos .unique() para obtener los valores únicos de la columna 'Provincia'
    poblacion_provincia = dataset[dataset['Provincia'] == provincia]['Poblacion'] # Filtramos los municipios de cada provincia
    plt.hist([poblacion_provincia], alpha = 0.5) # Usamos plt.hist para crear un histograma de la población de los municipios de cada provincia

plt.xlabel('Población') # Usamos plt.xlabel para poner la etiqueta del eje x
plt.ylabel('frecuencia') # Usamos plt.ylabel para poner la etiqueta del eje y
plt.legend(dataset['Provincia'].unique(), title='Provincias') # Usamos plt.legend para poner la leyenda del gráfico
plt.tight_layout()
plt.show()
```

#### Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
# Creamos una muestra al azar sin repetición con sample
muestra = dataset.sample(50, replace=False) # Usamos .sample para obtener una muestra al azar de 10 municipios

#Calculamos la media de la población de la muestra
media_muestra = muestra['Poblacion'].mean() # Calculamos la media de la población de la muestra con mean()
print(f"La media de la población de la muestra es: {media_muestra}")

#Calculamos la desviación estándar de la población de la muestra
desviacion_muestra = muestra['Poblacion'].std() # Calculamos la desviación estándar de la población de la muestra con std()
print(f"La desviación estándar de la población de la muestra es: {desviacion_muestra}")

#Imprimimos el resultado de los municipios de la muestra
print("Los municipios de la muestra son:")
print(muestra['Municipio'])
```
